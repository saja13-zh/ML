```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/players_22.csv")
data
```

```python
newData=data[["overall","potential","value_eur","wage_eur","age"]]
print(newData.head())
print(newData.info()) #there is missing data in "value_eur" and "wage_eur"
Data=newData.dropna() #remove missing value
print(Data.describe())
print(Data.info()) # now there is no missing value
#Scale data to standardize values
Data1 = ((Data - Data.min()) / (Data.max() - Data.min())) * 10 + 1
print(Data1)
Data1.describe() #the data in the same range(min,max)
Data=Data.values

fig = plt.figure(0)
plt.grid(True)
plt.scatter(Data[:,0],Data[:,1])
plt.show()
```

```python
from types import LambdaType
k=3 #the best number of cluster for thes data
#1- initialize random centroids
def rand_c(Data1,k):
  centroid=[]
  for i in range(k):
    c=Data1.apply(lambda x : float(x.sample()))
    centroid.append(c)

  centroid=pd.concat(centroid,axis=1)
  return centroid
```

```python
centroid=rand_c(Data1,k)
print(centroid)
```

```python
#2-label each data point into cluster
#x=centroid.iloc[:,0]
def ass(data,centroid):
  distance=centroid.apply(lambda x:np.sqrt(((data-x)**2).sum(axis=1)))# calculate the distance
  print(f"the distance btween each point and centroid = \n{distance} ")
#3-cluster assignment
  ass=distance.idxmin(axis=1)
  return ass
```

```python
data_label=ass(Data1,centroid)
print(f"data label \n{data_label}")
print(data_label.value_counts()) # info about each cluster
```

```python
#update the centroid
def update_c(Data1,data_label,k):
  g=Data1.groupby(data_label)
  update=g.apply(lambda x:np.mean(x)).T
  return update

print(update_c(Data1,data_label,k))
```

```python
#function for plot
from sklearn.decomposition import PCA
from IPython.display import clear_output

def plot(Data1,data_label,centroid,iteration):
  pca=PCA(n_components=2)
  data_2d=pca.fit_transform(Data1)
  c_2d=pca.transform(centroid.T)
  clear_output(wait=True)
  plt.title(f"iteration ={iteration}")
  plt.scatter(x=data_2d[:,0],y=data_2d[:,1],c=data_label)
  plt.scatter(x=c_2d[:,0],y=c_2d[:,1])
  plt.show()
```

```python
max_iteration=100
centroid=rand_c(Data1,k)
old_c=pd.DataFrame()
iteration=1
while iteration<max_iteration and not centroid.equals(old_c):
  old_c=centroid
  data_label=ass(Data1,centroid)
  centroid=update_c(Data1,data_label,k)
  plt_c=plot(Data1,data_label,centroid,iteration)
  iteration+=1
```
