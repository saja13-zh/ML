```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/players_22.csv")
data
```

```python
newData=data[["overall","potential","value_eur","wage_eur","age"]]
print(newData.head())
print(newData.info()) #there is missing data in "value_eur" and "wage_eur"
Data=newData.dropna() #remove missing value
print(Data.describe())
print(Data.info()) # now there is no missing value
#Scale data to standardize values
Data1 = ((Data - Data.min()) / (Data.max() - Data.min())) * 10 + 1
print(Data1)
Data1.describe() #the data in the same range(min,max)
Data
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

x1 = linkage(Data, method='ward', metric='euclidean')
plt.figure(figsize=(8, 6))
dendrogram(x1)
plt.title('Dendrogram')
plt.xlabel('Data Index')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(n_clusters=4, linkage='ward')
labels = clustering.fit_predict(Data)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Visualize the clusters
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Choose x, y, z axis for visualization
x = Data['potential']
y = Data['wage_eur']
z = Data['value_eur']

ax.scatter(x, y, z, c=labels, s=50)
ax.set_xlabel('Potential')
ax.set_ylabel('Wage (EUR)')
ax.set_zlabel('Value (EUR)')

plt.title('Agglomerative Clustering - FIFA 22 Players')
plt.show()
```

```python
data2=pd.read_csv("/content/countries_continents.csv")
data2=data2[["Longitude","Latitude"]]
data2
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

x1 = linkage(data2, method='single', metric='euclidean')
x2 = linkage(data2, method='complete', metric='euclidean')
x3 = linkage(data2, method='average', metric='euclidean')
x4 = linkage(data2, method='ward', metric='euclidean')

plt.figure(figsize=(15, 10))
plt.subplot(2,2,1), dendrogram(x1), plt.title('Single')
plt.subplot(2,2,2), dendrogram(x2), plt.title('Complete')
plt.subplot(2,2,3), dendrogram(x3), plt.title('Average')
plt.subplot(2,2,4), dendrogram(x4), plt.title('Ward')
plt.show()
```
